```python
# Python code to calculate descriptive statistics
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
```

## Exploratory Data Analysis (EDA)

```python
# Summary statistics for numerical columns
print(df.describe())

```

## Visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Python code for data visualization (e.g., histogram)
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title('Distribution of Age on Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

## Correlation Analysis

```python
# Python code for correlation matrix
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix['Survived'])
```

```python
# Display correlation of 'Fare' and 'Age' with 'Survived'
print("Correlation of Fare with Survived:", correlation_matrix['Survived']['Fare'])
print("Correlation of Age with Survived:", correlation_matrix['Survived']['Age'])
```

## Regression Modelling

```python
# Python code for a simple linear regression plot
sns.regplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare Regression')
plt.show()
```

## Creating a dataframe

### Subtask:
Generate a pandas DataFrame with 20 data points and the specified columns: 'name', 'age', 'department', and 'salary'.


**Reasoning**:
The subtask is to create a pandas DataFrame. A dictionary with the specified keys and lists of 20 data points, then create a DataFrame from it and display its head.



```python
import pandas as pd
import numpy as np

data = {
    'name': [f'Name_{i}' for i in range(20)],
    'age': np.random.randint(20, 60, 20),
    'department': np.random.choice(['Sales', 'IT', 'HR', 'Marketing'], 20),
    'salary': np.random.randint(30000, 100000, 20)
}

df_new = pd.DataFrame(data)
print(df_new.head())
```

## Perform eda

### Subtask:
Conduct basic exploratory data analysis on the generated DataFrame, including descriptive statistics and checking for missing values.


**Reasoning**:
Conduct basic exploratory data analysis by displaying descriptive statistics and checking for missing values in the `df_new` DataFrame.



```python
# Display descriptive statistics
print("Descriptive Statistics:")
display(df_new.describe())

# Check for missing values
print("\nMissing Values:")
print(df_new.isnull().sum())
```

## Visualize data

### Subtask:
Create different types of visualizations to explore the relationships and distributions within the data (e.g., histograms, bar plots, scatter plots).


**Reasoning**:
Generate the required visualizations based on the instructions, including a histogram of age, a bar plot of department counts, a scatter plot of age vs. salary, and a box plot of salary by department.



```python
# Histogram of Age
plt.figure(figsize=(8, 6))
sns.histplot(df_new['age'], bins=5, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of Department counts
plt.figure(figsize=(8, 6))
sns.countplot(x='department', data=df_new)
plt.title('Number of Employees per Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Scatter plot of Age vs. Salary
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='salary', data=df_new)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Box plot of Salary by Department
plt.figure(figsize=(8, 6))
sns.boxplot(x='department', y='salary', data=df_new)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()
```

## Perform analysis

### Subtask:
Based on the EDA and visualizations, perform relevant analysis, which could include grouping data, calculating aggregates, or identifying trends.


**Reasoning**:
Calculate the average salary and employee count per department and find the departments with the highest average salary and most employees.



```python
# Calculate the average salary for each department
average_salary_per_department = df_new.groupby('department')['salary'].mean()
print("Average Salary per Department:")
print(average_salary_per_department)
```

```python
# Find the department with the highest average salary
highest_average_salary_department = average_salary_per_department.idxmax()
print(f"\nDepartment with the Highest Average Salary: {highest_average_salary_department}")
```

```python
# Calculate the number of employees in each department
employee_count_per_department = df_new.groupby('department').size()
print("\nNumber of Employees per Department:")
print(employee_count_per_department)
```

```python
# Find the department with the most employees
most_employees_department = employee_count_per_department.idxmax()
print(f"\nDepartment with the Most Employees: {most_employees_department}")
```
